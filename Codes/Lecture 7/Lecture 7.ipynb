# Seaborn

## Importing library

```python
import seaborn as sns
```

## Import bill dataset

```python
print(sns.get_dataset_names())
data = sns.load_dataset('tips')
print(data)
```

## Distribution plot

```python
sns.distplot(data['total_bill'])
```

## Scatter plot

```python
sns.scatterplot(data['total_bill'])
```

## Joint plot

```python
sns.jointplot(data['total_bill'])
```

## Box plot

```python
sns.boxplot(data['total_bill'])
```

# Bar plot

```python
sns.barplot(x=data.day, y=data.total_bill)
```

## Pair plot

```python
sns.pairplot(data)
```

## Heat map

```python
data.corr()  #correlation matrix
```

```python
sns.heatmap(data.corr())
```
